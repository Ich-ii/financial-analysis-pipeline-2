```python
import sys
import pandas as pd

project_root = r"C:\Users\ADMIN\Documents\My Documents\MyDataAnalysis\Financial statement analysis\financial-analysis-pipeline"
sys.path.insert(0, project_root)

from orchestrator.orchestrator import afap_run

financials = pd.read_csv("../data/cleaned/financial_statements.csv")

results = afap_run(financials)

results.keys()
```

```python
for k, v in results.items():
    print(k, type(v))
```

```python
import pandas as pd

tables = {}

for key, value in results.items():
    if isinstance(value, pd.DataFrame):
        tables[key] = value
    elif isinstance(value, list):
        tables[key] = pd.DataFrame(value)
    elif isinstance(value, dict):
        tables[key] = pd.DataFrame([value])
```

```python
tables["ratios"]
tables["trend"]
tables["cash_flow"]
tables["solvency"]
tables["composite_risk"]
```

```python
summary = pd.concat(
    [
        tables["ratios"].assign(section="Ratios"),
        tables["trend"].assign(section="Trend"),
        tables["cash_flow"].assign(section="Cash Flow"),
        tables["solvency"].assign(section="Solvency"),
        tables["composite_risk"].assign(section="Composite Risk"),
    ],
    ignore_index=True
)

summary
```
