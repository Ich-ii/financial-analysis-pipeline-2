```python
import sys
import os
import pandas as pd

sys.path.append(os.path.abspath('..'))
from orchestrator.orchestrator import afap_run

financials = pd.read_csv("../data/cleaned/financial_statements.csv")

# Baseline
base_results = afap_run(financials)

# Conservative client
conservative_config = {
    "risk_weights": {
        "solvency": 50,
        "cash_flow": 30,
        "trend": 10,
        "anomaly": 10
    },
    "risk_bands": {
        "medium": 20,
        "high": 40
    }
}

conservative_results = afap_run(financials, conservative_config)
```
